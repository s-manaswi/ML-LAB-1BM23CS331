```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Canada Income
df1 = pd.read_csv("canada_per_capita_income.csv")
X1 = df1[['year']]
y1 = df1['per capita income (US$)']

model1 = LinearRegression()
model1.fit(X1, y1)

income_2020 = model1.predict(pd.DataFrame([[2020]], columns=['year']))
print("Predicted Per Capita Income in 2020: $", income_2020[0])


# Salary Prediction
df2 = pd.read_csv("salary.csv")
df2 = df2.dropna()

X2 = df2[['YearsExperience']]
y2 = df2['Salary']

model2 = LinearRegression()
model2.fit(X2, y2)

salary_12 = model2.predict(pd.DataFrame([[12]], columns=['YearsExperience']))
print("Predicted Salary for 12 Years Experience: $", salary_12[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset (just filename)
df = pd.read_csv('housing_area_price.csv')

print(df.head())

X = df[['area']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

predicted_price = model.predict(pd.DataFrame([[3300]], columns=['area']))

print("Predicted price for 3300 sq ft:", predicted_price[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('homeprices_Multiple_LR.csv')

# Fill missing values
df = df.fillna(df.mean())

print(df.head())

X = df[['area', 'bedrooms', 'age']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

predicted_price = model.predict(pd.DataFrame([[3000, 3, 40]],
                                             columns=['area', 'bedrooms', 'age']))

print("Predicted price:", predicted_price[0])
```

```python
import warnings
warnings.filterwarnings("ignore")   # 🔹 Hide all warnings

import pandas as pd
from sklearn.linear_model import LinearRegression


df1 = pd.read_csv('hiring.csv')

# Clean column names
df1.columns = df1.columns.str.strip()
df1.columns = df1.columns.str.replace(' ', '_', regex=False)
df1.columns = df1.columns.str.replace('(out_of_10)', '', regex=False)
df1.columns = df1.columns.str.replace('(', '', regex=False)
df1.columns = df1.columns.str.replace(')', '', regex=False)
df1.columns = df1.columns.str.replace('$', '', regex=False)

# Convert experience words to numbers
word_to_num = {
    'zero':0,'one':1,'two':2,'three':3,'four':4,
    'five':5,'six':6,'seven':7,'eight':8,
    'nine':9,'ten':10,'eleven':11,'twelve':12
}

df1['experience'] = df1['experience'].map(word_to_num).fillna(df1['experience'])
df1 = df1.apply(pd.to_numeric)
df1 = df1.fillna(df1.mean())

X1 = df1.iloc[:, :-1]
y1 = df1.iloc[:, -1]

model1 = LinearRegression()
model1.fit(X1, y1)

# 🔹 Use DataFrame to remove feature name warning
input1 = pd.DataFrame([[2,9,6]], columns=X1.columns)
input2 = pd.DataFrame([[12,10,10]], columns=X1.columns)

print("Predicted Salary (2 yr, 9 test, 6 interview):",
      model1.predict(input1)[0])

print("Predicted Salary (12 yr, 10 test, 10 interview):",
      model1.predict(input2)[0])


df2 = pd.read_csv('1000_Companies.csv')
df2 = pd.get_dummies(df2, columns=['State'], drop_first=True)

X2 = df2.iloc[:, :-1]
y2 = df2.iloc[:, -1]

model2 = LinearRegression()
model2.fit(X2, y2)

input_profit = pd.DataFrame(
    [[91694.48, 515841.3, 11931.24] + [0]*(X2.shape[1]-3)],
    columns=X2.columns
)

print("Predicted Profit:",
      model2.predict(input_profit)[0])
```
